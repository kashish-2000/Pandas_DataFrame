```python
import matplotlib.pyplot as plt

```

```python
data_frame1['Lat_N'].value_counts().plot(kind='bar')
plt.title('Bar Chart of Category Counts')
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.show()
```

```python
#plot line chart
data_frame1.plot(x='City', y='Lat_N', kind='line')
plt.title('Line Chart')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```

```python
#histogram chart1
data_frame1['Lat_N'].plot(kind='hist', bins=10, edgecolor='black')
plt.title('Histogram of Values')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# pie chart
category_counts = data_frame1['Lat_N'].value_counts()
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Category Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
